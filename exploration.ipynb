## Decrypting the Bible Verses

```python
class JWPDDescript:
    def __init__(self, content_key):
        dummy_hex = "11CBB5587E32846D4C26790C633DA289F66FE5842A3A585CE1BC3A294AF5ADA7"
        dummy = bytes.fromhex(dummy_hex)
        hashed_content_key = SHA256.new(content_key.encode()).digest()
        self.key = bytes(a ^ b for a, b in zip(dummy[:16], hashed_content_key[:16]))
        self.iv = bytes(a ^ b for a, b in zip(dummy[16:], hashed_content_key[16:]))

    def apply(self, protected_data):
        decrypted_data = self.decrypt(protected_data)
        if decrypted_data:
            return zlib.decompress(decrypted_data)
        return None

    def decrypt(self, data):
        cipher = AES.new(self.key, AES.MODE_CBC, iv=self.iv)
        decrypted = cipher.decrypt(data)
        return decrypted
```

```python
import re

# Adjusted pattern to match content more reliably within <span> tags
pattern = r'<span[^>]*>(.*?)<\/span>'

def get_content_regex(content):
    # Check if the content is a bytes-like object and decode it to a string if true
    if isinstance(content, bytes):
        content = content.decode('utf-8')
    
    # Find all matches of the pattern
    extracted_text = re.findall(pattern, content, re.DOTALL)  # re.DOTALL allows . to match newlines as well
    
    # Join all found texts instead of returning just the first one
    if extracted_text:
        return ' '.join(extracted_text).strip()  # Join and strip to remove leading/trailing whitespace
    return None

# Applying the adjusted function
bibleVerse_fr['content_extracted'] = bibleVerse_fr['content_decrypted'].apply(get_content_regex)
bibleVerse_fr['content_extracted']
```

```python
from bs4 import BeautifulSoup

def get_content_bs(content):
    # Check if the content is a bytes-like object and decode it to a string if true
    if isinstance(content, bytes):
        content = content.decode('utf-8')
    
    # Use BeautifulSoup to parse the HTML content
    soup = BeautifulSoup(content, 'html.parser')
    
    # Find all <span> tags and extract their text
    spans = soup.find_all('span')
    extracted_texts = [span.text for span in spans]
    
    # Join all extracted texts
    if extracted_texts:
        return ' '.join(extracted_texts).strip()
    return None

# Applying the BeautifulSoup function
bibleVerse_fr['content_extracted'] = bibleVerse_fr['content_decrypted'].apply(get_content_bs)
bibleVerse_fr['content_extracted']
```
